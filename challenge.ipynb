```python
#importing important libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv")
```

```python
# Checking if there is Star type column.
if 'Star type' in data.columns:
    # Cleaning the data to handle any missing or incorrect values
    data.dropna(inplace=True)
    data = data[data["Star type"] != "?"]

    # Convert star_type to categorical variable
    data["Star type"] = data["Star type"].astype('category')

    # converting star_type into multiple columns
    data = pd.get_dummies(data, columns=["Star type"])

    #finding the column with maximum sum which gives the most common star type
    most_common_star_type = data.filter(regex='Star type_').sum().idxmax().split("_")[-1]

    # Displaying the results
    print("The most common star type is:", most_common_star_type)
else:
    print("The 'Star type' column does not exist in the data")
    
```

```python
# Plotting luminosity against temperature
plt.scatter(data["Temperature (K)"], data["Luminosity(L/Lo)"])

# Adding labels to the plot
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity(L/Lo))")
plt.title("Hertzsprung-Russell Diagram")

# displaying the plot
plt.show()

```
